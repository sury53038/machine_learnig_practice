```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
num_transform = SimpleImputer(strategy = 'mean')
cat_transform = SimpleImputer(strategy = 'most_frequent')
```

```python
df.info()
```

```python
df['NumberOfPersonVisiting'] = df['NumberOfPersonVisiting'] + df['NumberOfChildrenVisiting']
```

```python
df = df.drop(columns = ['CustomerID', 'NumberOfChildrenVisiting'], axis = 1)
print(df.columns)
df.head()
```

```python
y = df['ProdTaken']
df = df.drop(['ProdTaken'], axis = 1)
df.head()
```

```python
numeric_cols = [cols for cols in df.columns if df[cols].dtype != 'O']
cat_cols = [cols for cols in df.columns if df[cols].dtype == 'O']
```

```python
from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
ohe = OneHotEncoder()
ord = OrdinalEncoder()
```

```python
df['TypeofContact'].unique()
```

```python
df['TypeofContact'] = df['TypeofContact'].fillna('Self Enquiry')
df['TypeofContact'].unique()
```

```python
df['Occupation'].unique()
```

```python
df['Gender']=df['Gender'].replace('Fe Male', 'Female')
df['Gender'].unique()
```

```python
df['ProductPitched'].unique()
```

```python
df['MaritalStatus']=df['MaritalStatus'].replace('Unmarried', 'Single')
df['MaritalStatus'].unique()
```

```python
df['Designation'].unique()
```

```python
ohe_cols = ['TypeofContact','Gender']
odn_cols = ['Occupation','ProductPitched','MaritalStatus','Designation']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size = 0.30, random_state = 42)
```

```python
processing1 = ColumnTransformer(
    [   
        ('fillnull', Pipeline([
            ('imputer', SimpleImputer(strategy = 'mean')),
            ('scaler', StandardScaler())
        ]), numeric_cols),

        ('ohe', Pipeline([
            ('imputer', SimpleImputer(strategy = 'most_frequent')),
            ('ohe', ohe)
        ]),ohe_cols),

        ('ord',Pipeline([
            ('imputer', SimpleImputer(strategy = 'most_frequent')),
            ('encoder', ord)
        ]),odn_cols),       
    ],remainder = 'drop'

)
```

```python
pipe = Pipeline(
    [
        ('processing', processing1)
    ]
)
```

```python
df.columns
```

```python
x_train_transformed = pipe.fit_transform(x_train)
x_test_transformed = pipe.fit_transform(x_test)
```

```python
print(x_train_transformed, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn. linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
models = {
    "LogisticRegression": LogisticRegression(),
    "AdaBoostClassifier": AdaBoostClassifier(),
    "GradientBoostingClassifier":GradientBoostingClassifier(),
    "RandomForestClassifier": RandomForestClassifier()
}
```

```python
for key,values in models.items():
    model = values
    model.fit(x_train_transformed, y_train)

    y_train_pred = model.predict(x_train_transformed)
    y_pred = model.predict(x_test_transformed)

    print(f"Training accuracy of {key}")
    model_training_accuracy = accuracy_score(y_train_pred, y_train)
    print("Accuracy Score :: ",model_training_accuracy)
    model_training_f1 = f1_score(y_train_pred, y_train)
    print("F1 Score :: ",model_training_f1)
    model_recall_score = recall_score(y_train_pred, y_train)
    print("Recall Score :: ",y_train_pred)
    model_precision_score = precision_score(y_train_pred,y_train)
    print("Precision Score ::", model_precision_score)
    print(f"\nTest accuracy of {key}")
    model_test_accuracy = accuracy_score(y_pred, y_test)
    print("Accuracy Score :: ",model_test_accuracy)
    model_test_f1 = f1_score(y_pred, y_test)
    print("F1 Score :: ",model_test_f1)
    model_recall_score = recall_score(y_pred, y_test)
    print("Recall Score :: ",model_recall_score)
    model_precision_score = precision_score(y_pred,y_test)
    print("Precision Score :: ",model_precision_score)

    print('='*35)
    print('\n')
```

### HyperParameterTuning

```python
rf_params = {
    "n_estimators":[100,200,300,500],
    "min_samples_split":[2,5,8,10],
    "max_depth":[5,8,12,None]
}
gradient_boost = {
    "n_estimators":[100,200,300,500],
    "learning_rate":[0.2,0.3,0.5,0.4],
    "max_depth":[2,3,4,5],
    "min_samples_split":[2,3,4,5],
    "max_features":['sqrt','log2']
}
```

```python
rf_params, gradient_boost
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv = [
    ("RandomForest", RandomForestClassifier(),rf_params),
    ("GradientBoost", GradientBoostingClassifier(), gradient_boost)
]
```

```python
x_test_transformed.shape
```

```python
model_params = {}
for name,model,param in randomcv:
    random = RandomizedSearchCV(estimator = model, param_distributions = param, n_iter = 100, cv = 3, verbose = 2, n_jobs = -1)
    random.fit(x_train_transformed, y_train)

    model_params[name] = random.best_params_

for model_name in model_params:
    print(f"\n\n----{model_name}----")
    print(model_params[model_name])
```

```python
modelss = {
    # "LogisticRegression": LogisticRegression(),
    "GradientBoostingClassifier": GradientBoostingClassifier(n_estimators= 500, min_samples_split= 2, max_features= 'log2', max_depth= 5, learning_rate= 0.4),
    # "GradientBoostingClassifier":GradientBoostingClassifier(),
    "RandomForestClassifier": RandomForestClassifier(n_estimators= 200, min_samples_split= 2, max_depth= None)
}
```

```python
for key,values in modelss.items():
    model = values
    model.fit(x_train_transformed, y_train)

    y_train_pred = model.predict(x_train_transformed)
    y_pred = model.predict(x_test_transformed)

    print(f"Training accuracy of {key}")
    model_training_accuracy = accuracy_score(y_train_pred, y_train)
    print("Accuracy Score :: ",model_training_accuracy)
    model_training_f1 = f1_score(y_train_pred, y_train)
    print("F1 Score :: ",model_training_f1)
    model_recall_score = recall_score(y_train_pred, y_train)
    print("Recall Score :: ",y_train_pred)
    model_precision_score = precision_score(y_train_pred,y_train)
    print("Precision Score ::", model_precision_score)
    print(f"\nTest accuracy of {key}")
    model_test_accuracy = accuracy_score(y_pred, y_test)
    print("Accuracy Score :: ",model_test_accuracy)
    model_test_f1 = f1_score(y_pred, y_test)
    print("F1 Score :: ",model_test_f1)
    model_recall_score = recall_score(y_pred, y_test)
    print("Recall Score :: ",model_recall_score)
    model_precision_score = precision_score(y_pred,y_test)
    print("Precision Score :: ",model_precision_score)

    print('='*35)
    print('\n')
```

```python
model_pipeline = Pipeline([
    ("first", processing1),
    ("second",GradientBoostingClassifier(n_estimators= 500, min_samples_split= 2, max_features= 'log2', max_depth= 5, learning_rate= 0.4))
])
```
