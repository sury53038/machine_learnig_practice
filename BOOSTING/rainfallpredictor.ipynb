```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'GradientBoost': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier()
}
```

```python
df['Location'] = df['Location'].replace('MelbourneAirport', 'Melbourne')
```

```python
df = df.drop(columns=['Date'])
```

```python
df.head()
```

```python
y = df['RainTomorrow']
df = df.drop(columns = ['RainTomorrow'], axis = 1)
df.head()
```

```python
numeric_cols = [cols for cols in df.columns if df[cols].dtype != 'O']
cat_cols = [cols for cols in df.columns if df[cols].dtype == 'O']
```

```python
numeric_cols
```

```python
cat_cols
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('Scaler',StandardScaler(), numeric_cols),
    ('Encoder',OneHotEncoder(), cat_cols)
], remainder= 'passthrough'
)
```

```python
x = df
x.head()
```

```python
y
```

```python
y =LabelEncoder().fit_transform(y)
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state= 42)
```

```python
x_train, x_test
```

```python
y_train, y_test
```

```python
x_train_transformed = preprocessor.fit_transform(x_train)
x_test_transformed = preprocessor.fit_transform(x_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, r2_score, precision_score,f1_score, classification_report
```

```python
for key, model in models.items():
    model.fit(x_train_transformed, y_train)
    
    # Predictions
    y_train_pred = model.predict(x_train_transformed)
    y_test_pred = model.predict(x_test_transformed)

    print(f"\n================ {key} Model ================\n")

    # --- Training Performance ---
    print("Training Performance:")
    print(f"Accuracy:  {accuracy_score(y_train, y_train_pred)}")
    print(f"Precision: {precision_score(y_train, y_train_pred)}")
    print(f"Recall:    {recall_score(y_train, y_train_pred)}")
    print(f"F1 Score:  {f1_score(y_train, y_train_pred)}")

    # --- Testing Performance ---
    print("\nTesting Performance:")
    print(f"Accuracy:  {accuracy_score(y_test, y_test_pred)}")
    print(f"Precision: {precision_score(y_test, y_test_pred)}")
    print(f"Recall:    {recall_score(y_test, y_test_pred)}")
    print(f"F1 Score:  {f1_score(y_test, y_test_pred)}")

    # Optional: Detailed Report
    # print("\nClassification Report (Test Data):")
    # print(classification_report(y_test, y_test_pred))

    print("===============================================\n")
```

```python
rf_params = {
    'n_estimators': [100,200,300],
    'criterion': ['gini','entropy','log_loss'],
    'max_depth': [5, 8, 10, None],
    'min_samples_split': [2, 4, 5]
}
xgb_params = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 300, 500],
    'colsample_bytree': [0.5, 0.8, 1],
    'subsample': [0.5, 0.8, 1],
    'gamma': [0, 1, 5]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv = [
    ('Random Forest', RandomForestClassifier(), rf_params),
    ('XGBClassifier', XGBClassifier(), xgb_params)
]
```

```python
model_params = {}
for name,model,params in randomcv:
    random = RandomizedSearchCV(estimator=model, param_distributions=params, n_jobs=-1,n_iter = 100, cv = 3, verbose = 2)
    random.fit(x_train_transformed, y_train)

    model_params[name] = random.best_params_

    print(f"\n\n.........{name} best params........")
    print(model_params[name])
```

```python
modelss = {
    'Random Forest': RandomForestClassifier(n_estimators= 300, min_samples_split= 5, max_depth= None, criterion= 'gini'),
    'XGBoost': XGBClassifier(subsample= 0.5, n_estimators= 500, max_depth= 7, learning_rate= 0.05, gamma= 1, colsample_bytree= 0.8)
}
```

```python
for key, values in modelss.items():
    model = values
    model.fit(x_train_transformed, y_train)

    y_pred = model.predict(x_test_transformed)

    print(f"- For {key} model")
    model_accuracy = accuracy_score(y_pred, y_test)
    print("...Accuracy Score: ", model_accuracy)
    model_precision = precision_score(y_pred, y_test)
    print("...Precision Score: ",model_precision)
    model_r2 = r2_score(y_pred, y_test)
    print("...R2 Score: ", model_r2)
    model_recall = recall_score(y_pred, y_test)
    print("...Recall Score: ", model_recall)
```

```python
from sklearn.pipeline import Pipeline
```

```python
model_pipe = Pipeline([
    ('first', preprocessor),
    ('second',XGBClassifier(subsample= 0.8, n_estimators= 500, max_depth= 7, learning_rate= 0.1, gamma= 0, colsample_bytree= 0.5))
])
```

```python
model_pipe.fit(x_train, y_train)
```
