```python
df['PCOS_Diagnosis'] = df['PCOS_Diagnosis'].astype('category')
print(df.dtypes)
```

```python
df.columns
```

```python
features = ['Age', 'BMI', 'Menstrual_Irregularity', 'Testosterone_Level(ng/dL)',
       'Antral_Follicle_Count']
```

```python
X = df[features]
```

```python
Y = df['PCOS_Diagnosis']
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
Y = LabelEncoder().fit_transform(Y)
```

```python
x_train, y_train, x_test, y_test = train_test_split(X,Y, test_size = 0.25, random_state = 42)
```

```python
x_train
```

```python
x_test, y_test
```

```python
scaler = StandardScaler()
# x_train_reshape = x_train.reshape(-1,1)
# y_train_reshape = y_train.reshape(-1,1)
x_train = scaler.fit_transform(x_train)
y_train = scaler.fit_transform(y_train)
```

```python
y_train
```

```python
dt_clf = DecisionTreeClassifier()
```

```python
dt_clf.fit(x_train, x_test)
```

```python
y_pred = dt_clf.predict(y_train)
```

```python
accuracy =  accuracy_score(y_test, y_pred)
```

```python
print("accuracy of the model :: ",accuracy)
```
